```python
mapping={"entailment":0,"neutral":1,"contradiction":2}

csv2=csv2.replace(to_replace="entailment",value=0)
csv2=csv2.replace(to_replace="neutral",value=1)
csv2=csv2.replace(to_replace="contradiction",value=2)

val_csv2 =val_csv2.replace(to_replace="entailment",value=0)
val_csv2 =val_csv2.replace(to_replace="neutral",value=1)
val_csv2 =val_csv2.replace(to_replace="contradiction",value=2)

test_csv2=test_csv2.replace(to_replace="entailment",value=0)
test_csv2=test_csv2.replace(to_replace="neutral",value=1)
test_csv2=test_csv2.replace(to_replace="contradiction",value=2)
```

```python
csv2.reset_index(inplace=True)
val_csv2.reset_index(inplace=True)
test_csv2.reset_index(inplace=True)
```

```python
csv2
```

```python
import torch
```

```python
csv_labels=csv2.loc[:,["label1","label2","label3","label4","label5"]]
l_probas=[p1,p2,p3,p4,p5]

probas_finale=[]
a=0
for i in range(len(csv_labels)):
    proba=[0.0,0.0,0.0]
    for j in range(5):
        index=csv_labels.iloc[i,j]
        proba[index]+=l_probas[j][index]
    probas_finale.append(proba/sum(proba))
    
    
csv2=csv2.replace(to_replace="-",value=9)
actual=list(csv2["gold_label"].astype({"gold_label":int}))
#list(map(int,val_data["labels"]))

predicted=torch.argmax(torch.Tensor(probas_finale),dim=-1).tolist()
#list(map(int,val_preds))

confusion_matrix = metrics.confusion_matrix(actual, predicted)

print(confusion_matrix)

csv2.insert(0,"probas_label",probas_finale)

csv2
```

```python
val_csv_labels=val_csv2.loc[:,["label1","label2","label3","label4","label5"]]

probas_finale=[]

for i in range(len(val_csv_labels)):
    proba=[0.0,0.0,0.0]
    for j in range(5):
        index=val_csv_labels.iloc[i,j]
        proba[index]+=l_probas[j][index]
    probas_finale.append(proba/sum(proba))
    
    
val_csv2=val_csv2.replace(to_replace="-",value=9)
actual=list(val_csv2["gold_label"].astype({"gold_label":int}))
#list(map(int,val_data["labels"]))

predicted=torch.argmax(torch.Tensor(probas_finale),dim=-1).tolist()
#list(map(int,val_preds))

confusion_matrix = metrics.confusion_matrix(actual, predicted)

print(confusion_matrix)

val_csv2.insert(0,"probas_label",probas_finale)

val_csv2
```
